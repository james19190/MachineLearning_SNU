#### 1. Conditional Statement

````python
# if else statement

grade = 'B'
if grade == 'A': # : important
  print('Grade: ',grade)
  print('Good job')
else: # : important
  print('Grade: ',grade)
  print('Try hard')
````

````python
SBP = 120 # systolic blood pressure
if SBP >= 120:
  print('Systolic blood pressure: ', SBP)
  print('In danger')
elif SBP >= 100:
  print('Systolic blood pressure: ', SBP)
  print('Careful')
else: # the last elif -> else
  print('Healthy')
````

````python
# SBP, DBP (diastolic blood pressure)
# if SBP >= 120 and DBP >= 80, then diagnosed with hypertension

SBP = 100
DBP = 60

if SBP >= 120 and DBP >= 80:
  print('Hypertension')
elif SBP >= 120 or DBP >= 80:
  print('In danger')
else:
  print('Healthy')
````

#### 2. Loop (Iteration)
- For
- While

````python
# For

A = [10,20,30,40,50]
for i in A:
  print(i)
````

````python
for i in range(10): # range(10) == [0,1,2,3,4,...,9]
  print(i)
````

````python
for idx, ele in enumerate(A):
  print('{}th element: {}'.format(idx+1,ele))
````

````python
# While

i = 0
while i < 10:
  i += 1
  print(i,'printed')
````

````python
i = 0
while i >= 0 and i < 10:
  i += 1
  print(i)
````

````python
# infinite loop ctrl+c

while 1: # True: / i >= 0
  i += 1
  print(i)
````
