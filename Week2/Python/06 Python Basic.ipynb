#### 1. Calculation - Numpy

- Complex functions (e.g., sine, exponential)
- Computation algorithms (e.g., matrix computation, calculus)
- Random number generation (e.g., numbers normally distributed)

```python
import numpy as np

x = np.linspace(1,11,11) # 11 elements from 1 to 11 with equal space
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.exp(x)
y_4 = np.log(x)

print(x)
print(y_1)
print(y_2)
print(y_3)
print(y_4)
```

```python
np.random.choice(x,3,replace=False)
```

```python
# Integer

a = np.random.randint(5) # from 0~4 (range)
print(a)
a_1 = np.random.randint(-3,5) # from -3 to 4 (range)
print(a_1)
a_2 = np.random.randint(-10,5,size=(2,4)) # (range; a to b-1), (size=)
print(a_2)
```

```python
# Numbers normally distributed

b = np.random.normal(1,1,size=(2,4)) # (mean), (standard deviation), (size=)
print(b)
print(b.shape)
b_1 = np.random.normal(1,1,5)
print(b_1)
```

```python
# Numbers standard normal distributed

c = np.random.randn(2,3) # (size)
print(c)
```

```python
# Uniform numbers

d = np.random.rand(2,5) # (size) from 0 to 1
print(d)
d_1 = np.random.rand(3)
print(d_1)
```

## Q. Normalize and visualize by specifying a column

+ Draw a histogram for the median_income column with bins 300
+ In the california_housing_train data, subtract the average value from the median_income column, divide its standard deviation, and assign it to the new median_income_norm column.
+ Draw a histogram for the converted median_income_norm column with bins 300

```python
plt.hist(df['median_income'],bins=300)
```

```python
# In the california_housing_train data, subtract the average value from the median_income column, divide its standard deviation, and assign it to the new median_income_norm column.

mean = df['median_income'].mean()
sd = df['median_income'].std()

df['median_income_norm'] = ((df['median_income'] - mean) / sd)
```

```python
plt.hist(df['median_income_norm'],bins=300)
```
