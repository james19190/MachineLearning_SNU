## **Class**
#### Syntax suitable for expressing one data with various properties

E.g. 1) Calculator
- Functions
 - Four basic operations
 - Result Display
 - Saving Intermediate values
 - Initialization

```python
class Calculator():
    def __init__(self):
        self.memory = 0
    def show(self):
        print('The value currently saved in this calculator: ', format(self.memory))
    def clear(self):
        self.memory = 0
    def add(self,a):
        self.memory += a
    def sub(self,a):
        self.memory -= a
    def mult(self,a):
        self.memory *= a
    def div(self,a):
        try:
            self.memory /= a
        except:
            print('impossible to divide into 0')
```

```python
cal_1 = Calculator()
cal_1.add(1)
cal_1.show()
```

```python
cal_1.clear()
cal_1.show()
```

```python
cal_2 = Calculator()
cal_2.sub(5)
cal_2.show()
```

```python
cal_2.add(10)
cal_2.show()
```

```python
cal_2.div(0)
```

## **Inheritance**
#### Parent (Super, Base) Class and Child (Sub, Derived) Class



E.g. 2) Advanced Calculator
- Linear equation: $ax+b$
- Quadratic equation: $ax^2+bx+c$

```python
class Advanced_Calculator(Calculator):
    def __init__(self):
        super(Advanced_Calculator,self).__init__()

    def linear_sol(self,a,b):
        self.memory = a*self.memory+b

    def quad_sol(self,a,b,c):
        self.memory = a*self.memory**2+b*self.memory+c
```

```python
adv_cal = Advanced_Calculator()
adv_cal.add(2)
adv_cal.show()
```

```python
adv_cal.linear_sol(8,1)
adv_cal.show()
```

```python
adv_cal.clear()
adv_cal.add(1)
adv_cal.quad_sol(2,4,1)
adv_cal.show()
```

## Q. Inherit Calculator and define a new class that calculates the sum and product of positive integers up to k and print some results.

```python
# Q. Inherit Calculator and define a new class that calculates the sum and product of positive integers up to k and print some results.
class K_Calculator(Calculator):
    def __init__(self):
        super(K_Calculator, self).__init__()

    def sum_k(self,k):
        self.memory = sum(range(1,k+1))
        print('The sum of positive integers up to {} is {}'.format(k,self.memory))

    def product_k(self,k):
        self.memory = 1
        for i in range (1,k+1):
            self.memory *= i
        print('The product of positive integers up to {} is {}'.format(k,self.memory))
    
```
