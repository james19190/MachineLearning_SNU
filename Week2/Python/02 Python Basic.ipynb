#### 4. Set
  - List
  - Dictionary

```
# List
a = [] # initialization
type(a)
```

```
a.append(1)
a.append('hey')
```

```
print(len(a))
print(a)
```

```
print(a[0]) # first component of list a
print(a[1]) # second component of list a
```

```
# Dictionary
a = {}
type(a)
```

```
# relation between key and lock / keys(), values() / items()
a['1'] = 0
a[0] = 'Hello'
```

```
print(a)
```

```
print(a[str(1)])
```

```
print(a.keys())
print(a.values())
```

```
a.items()
```
