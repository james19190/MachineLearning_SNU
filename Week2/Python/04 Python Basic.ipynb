E.g. 2) Solution of Quadratic Equation

$x$ satisfying $ax^2+bx+c=0$; $x = {{-b\pm{\sqrt {b^2-4ac}}}\over{2a}}$

(input) parameters: $a, b, c$

(output) result: $x(={{-b\pm{\sqrt {b^2-4ac}}}\over{2a}})$

- Case 1: ${b^2-4ac} > 0$
- Case 2: ${b^2-4ac} = 0$
- Case 3: ${b^2-4ac} < 0$

````
def quad_sol(a,b,c):
  if a == 0:
    print('Can\'t be defined')
    return
  else:
    disc = b**2-4*a*c
    if disc > 0:
        return (-b+disc**0.5)/(2*a), (-b-disc**0.5)/(2*a)
    elif disc == 0:
        return -b/(2*a)
    else:
        print('No solution')
        return
````

````
print(quad_sol(1,2,0))
print(quad_sol(1,2,1))
print(quad_sol(1,2,2))
print(quad_sol(0,1,2))
````
