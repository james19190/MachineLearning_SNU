#### 1. Elementary Arithmetic
 - +, -, \*, /, %, //, **



```python
a = 359-358
b = 259+1
c = 18*19
d = 100/2
a_1 = 7//4 # 7=4*1+3
a_2 = 7%4 # 7=4*1+3
a_3 = 2**4
```

```python
print(a)
print(b)
print(c)
print(d)
print(a_1)
print(a_2)
print(a_3)
```

#### 2. Assignment Operators
 - =, +=, -=, \*=, /=, %=, //=, **=

```python
a = 10
print(a)
```

```python
a = a+1
print(a)
```

```python
a += 1
print(a)
```

```python
a //= 2
print(a)
```

#### 3. Comparison Operators
 - ==, !=, >, < , >=, <=
 - True/False; Boolean

```python
a = 1
a == 1
print(a)
```

```python
b = 1==1
print(b)
```

```python
1 != 1
```

```python
3 >= 2
```
