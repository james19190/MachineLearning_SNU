#### 1. Univariate function
- $y=x^3+0.5x^2-0.7x+0.1$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Sample Generation

x = np.linspace(-5,5,200)
y = x**3+0.5*x**2-0.7*x+0.1
plt.plot(x,y)
plt.grid()
```

```python
# Numerical method to compute the first derivative

dy_dx = np.gradient(y,x)
plt.plot(x,dy_dx)
plt.grid()
```

```python
# Analytical method

dy_dx_a = 3*x**2+x-0.7

plt.figure(1)
plt.plot(x,dy_dx,'r*-', label='Numerical')
plt.plot(x,dy_dx,'b*-', label='Analytic')
plt.grid()
plt.legend()
```

#### 2. Multivariate function

- $z = f(x_1,x_2)$
 - Partial derivative in terms of $x_1$ and $x_2$:
   ${\partial{z}}\over{\partial{x_1}}$, ${\partial{z}}\over{\partial{x_2}}$

   ${{\partial{z}}\over{\partial{x_1}}} = \lim\limits_{h\to{0}}{{f(x_1+h,x_2)-f(x_1,x_2)}\over{h}}$

   ${{\partial{z}}\over{\partial{x_2}}} = \lim\limits_{h\to{0}}{{f(x_1,x_2+h)-f(x_1,x_2)}\over{h}}$
- $f(x_1,x_2)=x_1x_2 + 3x_1^2 - 0.5x_2 $
 - Partial derivative in terms of $x_1$ and $x_2$


```python
x1 = np.linspace(-3,3)
x2 = np.linspace(-3,3)

def func(x1,x2):
    return x1*x2+3*x1**2-0.5*x2

def grad_x1(x1,x2,ep):
    return (func(x1+ep,x2)-func(x1,x2))/ep

def grad_x2(x1,x2,ep):
    return (func(x1,x2+ep)-func(x1,x2))/ep

def pd_x1(x1,x2):
    return x2+6*x1

def pd_x2(x1,x2):
    return x1-0.5
```

```python
val = x2
val = 5*np.ones(x2.shape[0])
ep = 1e-10

grad1 = grad_x1(x1,val,ep)
grad2 = grad_x2(val,x2,ep)

plt.figure(1)
plt.plot(x1,grad1,'b-')
plt.plot(x2,grad2,'r-')

pd1 = pd_x1(x1,val)
pd2 = pd_x2(val,x2)

plt.figure(2)
plt.plot(x1,pd1,'b-')
plt.plot(x2,pd2,'r-')
```
